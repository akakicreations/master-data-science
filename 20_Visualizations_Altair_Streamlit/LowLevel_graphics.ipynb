```
# Youtube 
data = [ 
{ "followers": "100k-500k", "value": 12500 }, 
{ "followers": "500k-1m", "value": 25000 }, 
{ "followers": "1m-3m", "value": 125000 }, 
{ "followers": "3m-7m", "value": 187500 }, 
{ "followers": "over 7m", "value": 300000 } 
] 
data_df = pd.DataFrame(data) 


width_svg=500 
height_svg=150 
d = draw.Drawing(width_svg,height_svg) 

def scaleY(n): 
  len_data = len(data_df) 
  return n * height_svg/len_data 

def scaleW(n): 
  max_data = data_df["value"].max() 
  return n * width_svg/max_data 

d.append(draw.Text("Ingresos de Youtubers",12,10,height_svg-20))
for i,row in data_df.iterrows():
  x = 10 
  y = scaleY(i) 
  w = scaleW(row["value"]) 
  h = 10 
  color = "blue" 
  d.append(draw.Rectangle(x,y,w,h,fill=color)) 

d
```
