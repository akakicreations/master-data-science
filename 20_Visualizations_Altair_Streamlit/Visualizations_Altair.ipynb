```
world_temp = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/worldTemperatures.csv",header=None)
```

```
world_temp.columns.astype(str)
```

```
world_temp.columns = ['0', '1', '2', 'date', '4', 'temp', '6', '7', '8', '9', '10', 'lat', 'lon',
       'city', 'country']
```

```
world_temp.head()
```

```
alt.data_transformers.disable_max_rows() #Para deshabilitar un máximo

alt.Chart(world_temp).mark_bar().encode(
    x= 'country',
    y= 'mean(temp)'
)
 
# Así no funciona porque son muchos datos
```

```
temp = world_temp.groupby("country").mean().reset_index() 

alt.Chart(temp).mark_bar().encode( 
x=alt.X("country",sort=alt.Sort(field="temp",op="mean",order="descending")), 
y="mean(temp)" 
)
```

```
temp = world_temp.groupby("city").mean().reset_index() 

alt.Chart(temp).mark_bar().encode( 
x=alt.X("city",sort=alt.Sort(field="temp",op="mean",order="descending")), 
y="mean(temp)",
tooltip='city' 
).properties(
    width=600
)
```

```
temp = world_temp.groupby("date").mean().reset_index() 

# Le decimos cómo leer la fecha a altari, si no, sale mal
world_temp['date'] = pd.to_datetime(world_temp['date'], format='%Y%m%d')

alt.Chart(temp).mark_bar().encode( 
x=alt.X("date:T",sort=alt.Sort(field="temp",op="mean",order="descending")), 
y="mean(temp)" 
).properties(
    width=600
)
```

```
select_country = alt.selection(type="single",encodings=["x"]) 

temp = world_temp.groupby(["country","date"]).mean().reset_index() 

temp_bar = alt.Chart(temp).mark_bar().encode( 
x=alt.X("country",sort=alt.Sort(field="temp",op="mean",order="descending"),axis=None), 
y="mean(temp)", 
tooltip="country" 
).properties( 
width=600 
).add_selection(select_country) 

temp_line = alt.Chart(temp).mark_line().encode( 
x=alt.X("date:T"), 
y="mean(temp)", 
).properties( 
width=600 
).transform_filter( 
select_country 
) 

temp_bar&temp_line
```

```
# Ir a http://geojson.io/#map=2/20.0/0.0 donde se podría construir el mapa
# O buscar un mapa ya construido en formato geojson: https://github.com/datasets/geo-countries

##### ESTO SON LOS PUNTOS

temp = world_temp.groupby("city").mean().reset_index()
points = alt.Chart(temp).mark_circle().encode(
    latitude="lat", 
    longitude="lon",
    tooltip="city",
    color= alt.Color('mean(temp)', scale=alt.Scale(domain=[-20,10,40], range=['lightblue', 'orange', 'red']))
).properties(
        width= 1000,
        height= 700,
    )

##### ESTO VA A SER EL FONDO

# remote geojson data object
url = "https://raw.githubusercontent.com/datasets/geo-countries/master/data/countries.geojson" 
data_geojson_remote = alt.Data(url=url, format=alt.DataFormat(property='features',type='json')) 

# chart object 
background = alt.Chart(data_geojson_remote).mark_geoshape(
    fill='lightgrey',
    stroke = 'white'
).encode( 
color="properties.name:N" 
).properties( 
    width= 1000,
    height= 700,
projection={'type': 'identity', 'reflectY': True} 
)

background+points
```

```
select_snake = alt.selection(type="single",encodings=["x"]) 

scatter_snake = alt.Chart(snake).mark_circle().encode( 
    x='evidence', 
    y='popularity', 
    color='type', 
    ).properties( 
    width=700, 
    height=500, 
    title="Scatter snake" 
).interactive().transform_filter( 
select_snake
)

bar_snake = alt.Chart(snake).mark_bar().encode( 
    x='type', 
    y='count()', 
    color='type', 
    ).properties( 
    width=700, 
    height=500, 
    title="Bar snake" 
).interactive().add_selection(
  select_snake
) 

scatter_snake|bar_snake
```
