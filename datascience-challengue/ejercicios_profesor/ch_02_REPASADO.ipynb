```python
#Ahora añadimos el chunksize
bi=pd.read_csv('dataset_original/bookings.csv.bz2', sep="^", usecols=['arr_port','pax','year'], nrows=9999, chunksize=3000)
```

```python
# Aplicamos a todo el código
bi=pd.read_csv('dataset_original/bookings.csv.bz2', sep="^", usecols=['arr_port','pax','year'], nrows=9999, chunksize=3000)
all_chunks=pd.DataFrame()
for i, b in enumerate(bi):
    print(i)
    b=b.dropna()
    b=b[ b['year']== 2013]
    del b['year']
    b_gr=b.groupby(['arr_port']).sum().sort_values('pax', ascending=False).reset_index()
    #b_gr.reset_index(inplace=True)
    all_chunks=all_chunks.append(b_gr)
    
all_result=all_chunks.groupby(['arr_port']).sum().sort_values('pax', ascending=False).reset_index()   
#b_gr.head(10)
#Como se ve a la vez que se suma se ordena por valor de pasajeros
```

```python
all_result.head(10)
```

```python
%%time
bi=pd.read_csv('dataset_original/bookings.csv.bz2', sep="^", usecols=['arr_port','pax','year'], chunksize=1000000)
all_chunks=pd.DataFrame()
for i, b in enumerate(bi):
    print(i)
    b=b.dropna()
    b=b[ b['year']== 2013]
    del b['year']
    b_gr=b.groupby(['arr_port']).sum().sort_values('pax', ascending=False).reset_index()
    #b_gr.reset_index(inplace=True)
    all_chunks=all_chunks.append(b_gr)
    
all_result=all_chunks.groupby(['arr_port']).sum().sort_values('pax', ascending=False).reset_index()
#Tiempo que va a tardar en hacer todo el dataframe en bloques de 10000000, es la idea del chunksize,
# por eso el análisis es de 10 bloques, según dice el contador
```
