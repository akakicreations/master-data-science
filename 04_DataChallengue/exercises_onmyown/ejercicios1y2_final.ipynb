```python
#Vamos a probar todo el proceso junto CON LA MUESTRA

#Crear archivo de muestra otra vez, porque si no, nos da error
! cat bookings.csv | head -10000 > ./bookings.muestra.csv

import pandas as pd
import numpy as np
pd.options.display.max_columns = 40
pd.options.display.max_rows = 70

#Leemos el archivo
df_muestra = pd.read_csv('bookings.muestra.csv', sep="^", usecols=['arr_port','pax','year'], nrows=9999)

# Quitamos los espacios de la fila aeropuertos
df_muestra['arr_port'] = df_muestra['arr_port'].str.strip()

#Quitamos los espacios a la fila de columnas con la función creada y renombro la fila columns
from quit_espac import quit_espac
df_muestra_firstcolumn = quit_espac(df_muestra.columns.tolist())
df_muestra.columns = df_muestra_firstcolumn

#Quitamos los NaN
df_muestra = df_muestra.dropna(subset=['pax'])

#Nos quedamos con las filas del año 2013
df_muestra = df_muestra[ df_muestra['year']== 2013]

#La borramos después del filtrado
del df_muestra['year']

#Agrupamos por año, sumamos los valores y ordenamos de pax
df_muestra_grup = df_muestra.groupby(['arr_port']).sum().sort_values('pax', ascending=False)

#Reseteamos el índice
df_muestra_grup.reset_index(inplace=True)

#probamos, SI, FUNCIONA!!!!!!
df_muestra_grup.head(10)
```

```python
#Vamos a ver si funciona con el dataset completo
import pandas as pd
import numpy as np
pd.options.display.max_columns = 40
pd.options.display.max_rows = 70

df_muestra = pd.read_csv('bookings.csv', sep="^", usecols=['arr_port','pax','year'])
df_muestra['arr_port'] = df_muestra['arr_port'].str.strip()

from quit_espac import quit_espac
df_muestra_firstcolumn = quit_espac(df_muestra.columns.tolist())
df_muestra.columns = df_muestra_firstcolumn

df_muestra = df_muestra.dropna(subset=['pax'])
df_muestra = df_muestra[ df_muestra['year']== 2013]
del df_muestra['year']

df_muestra_grup = df_muestra.groupby(['arr_port']).sum().sort_values('pax', ascending=False)
df_muestra_grup.reset_index(inplace=True)
df_completo = df_muestra_grup
df_completo.head(10)

#Sí, funciona!!!
```
