### La función de decisión
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

```python
draw(X,y)
```

```python
# Para hacer una predicción: es importante que las formas de la muestra a predecir sean iguales que las de entrenamiento
sample = np.array([2,0]).reshape(-1,2) #Y así configuramos los datos y redimensinamos para que valga como muestra
model.predict(sample) #Dice que para la muestra, da resultado es 0, los puntos negros.
```

```python
sample = np.array([-2,2]).reshape(-1,2)
model.predict(sample) #Da ahora con otra muestra 1, los blancos
```

```python
# El método Decision da el valor de la función para el valor en cuestión
sample = np.array([2, 0]).reshape(1, -1)
model.decision_function(sample)
```

```python
sample = np.array([-2, 2]).reshape(1, -1)
model.decision_function(sample)
```

```python
#Para dibujar las muestras que hemos puesto antes
fig, ax = draw(X, y, show = False)
ax.scatter([2], [0], c = "Red")
ax.scatter([-2], [2], c = "Blue")
plt.show()
```

###### Calculemos la función de decisión para otros puntos

###### Función de decisión para una matriz de puntos

```python
plt.figure(figsize = (10, 8))
plt.scatter(X[:, 0], X[:, 1])
for cx in range(-3, 3):
    for cy in range(-2, 3):
        sample = np.array([cx, cy]).reshape(1, -1)
        value = model.decision_function(sample)[0]
        plt.text(cx, cy, round(value, 2), fontsize = 10)
plt.ylim(-2.5, 3.5)
plt.xlim(-3.5, 3.5)
plt.show()
#Así vemos los valores numéricos donde se situan
```

```python
colors = ["salmon", "lightgreen"]
plt.figure(figsize = (10, 8))
for cx in np.linspace(-3, 2, 25):
    for cy in np.linspace(-2, 2, 25):
        sample = np.array([cx, cy]).reshape(1, -1)
        value = model.predict(sample)[0]
        plt.scatter(cx, cy, c = colors[value])
plt.show()
# Así vemos las predicciones que hace para esos valores para cada número.
# Esto es hacerlo cutren porque se utiliza la función pcoloremesh
```

```python
xx = np.linspace(-3, 3, 200)
yy = np.linspace(-2, 3, 200)
XX, YY = np.meshgrid(xx, yy)
```

```python
ZZ = model.decision_function(np.c_[XX.ravel(), YY.ravel()]).reshape(XX.shape)
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sc = ax.pcolormesh(XX, YY, ZZ)
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = "autumn_r")
fig.colorbar(sc)
plt.show()
#Calcula cuáles son los valroes que hacen que sea 0 o 1 de forma gradual y no por puntos
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = "winter")
ax.contour(XX, YY, ZZ, levels = 10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sc = ax.pcolormesh(XX, YY, ZZ)
ax.contour(XX, YY, ZZ, levels = 10, cmap = "autumn")
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = "autumn_r")
fig.colorbar(sc)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.contourf(XX, YY, ZZ, levels = 10)
#ax.scatter(X[:, 0], X[:, 1], c = y, cmap = "autumn_r")
plt.show()
```

```python
ZZ = model.predict(np.c_[XX.ravel(), YY.ravel()]).reshape(XX.shape)
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sc = ax.pcolormesh(XX, YY, ZZ)
ax.scatter(X[:, 0], X[:, 1], c = y, cmap = "autumn_r")
plt.show()
```

# Kernels
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

<div style="background:#0099cc;padding:20px;color:#ffffff">
<strong>EJERCICIO</strong>
<div>Aplica los kernels vistos a un clasificador aplicado a las características "sepal_length" y "sepal_width" conjunto Iris al completo (considerando las tres clases).</div>
</div>

```python
iris = pd.read_csv("iris.csv")
iris.head()
iris = iris.reindex(["sepal_length", "sepal_width", "species"], axis = 1)
iris = iris.sample(frac = 1) #Lo desordeno
y = iris.pop("species").astype("category").cat.codes.values #El texto lo convierto en dato numerico
X = iris.values
```

```python
draw(X, y);
```

### Linear kernel

```python
model = SVC(kernel = "linear")
model.fit(X, y)
draw3(X, y, model);
```

### Polynomial kernel

```python
model = SVC(kernel = "poly", degree = 1, gamma = 1, coef0 = 0) #Con estos parámetros hacemos lo mismoq ue el kernel linear
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 2)
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 3)
model.fit(X, y)
draw3(X, y, model);
```

```python
model = SVC(kernel = "poly", degree = 3, C = 10)
model.fit(X, y)
draw3(X, y, model);
```

### Radial Basis Function (Gaussian)

Si se asigna a &gamma; el valor "auto", tomará el valor 1/n, siendo n el número de características.

```python
model = SVC(kernel = "rbf", gamma = "auto")
model.fit(X, y)
draw3(X, y, model);
```

Si &gamma; es muy pequeño, la función tenderá a comportarse linealmente:

```python
model = SVC(kernel = "rbf", gamma = 0.01)
model.fit(X, y)
draw3(X, y, model);
```

Y si es muy grande, tenderá al sobreentrenamiento:

```python
model = SVC(kernel = "rbf", gamma = 10)
model.fit(X, y)
draw3(X, y, model);
```

### Sigmoid kernel

```python
gammas = [0.001, 0.01, 0.1, 1]
for g in gammas:
    print(g)
    model = SVC(kernel = "sigmoid", gamma = g, coef0 = 0)
    model.fit(X, y)
    draw3(X, y, model);
```

# Kernels personalizados
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

```python
iris = sns.load_dataset("iris")
iris = iris[iris.species.isin(["setosa", "virginica"])]
iris = iris.loc[:, ["sepal_length", "sepal_width", "species"]]
iris = iris.sample(frac = 1)
y_iris = iris.pop("species").astype("category").cat.codes.values
X_iris = iris.values
```

```python
draw(X_iris, y_iris);
```

```python
def custom_kernel(X, Y):
    M = np.array([[2, 0], [0, 1]])
    return np.dot(np.dot(X, M), Y.T)
```

```python
model = SVC(kernel = custom_kernel)
model.fit(X_iris, y_iris)
```

# Escenarios de regresión
<div style = "float:right"><a style="text-decoration:none" href = "#inicio">Inicio</a></div>

```python
#El SVM se puede aplciar a regresión (no sólo a clasificación)
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
y = pd.Series(boston.target)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

###### Aplicamos hold-out

```python
from sklearn.model_selection import train_test_split
```

###### Importamos la clase SVR, instanciamos el algormit y lo entrenamos

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

from sklearn.svm import SVR
```

```python
model = SVR()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

###### Lo reentrenamos con los parámetros C = 1000 y epsilon = 1

```python
model = SVR( C= 1000, epsilon = 1)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
