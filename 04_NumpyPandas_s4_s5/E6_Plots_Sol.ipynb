# Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

___

## Plots

- Crea dos array (x e y):
    - x: 1000 números equiespaciados entre 0 a 3$\pi$
    - y: Resultado de aplicar la función $f(x) = \frac{1}{2}\sin^2(10 x) + \cos(x)$

```python
x = np.linspace(0, 3*np.pi, 1000)
```

```python
def f(x):
    return 1/2*np.sin(10*x)**2 + np.cos(x)
```

```python
y = f(x)
```

- Crea un gráfico de y en función de x.

```python
plt.plot(x, y)
```

- Realiza el mismo gráfico anterior pero con una línea discontínua y verde

```python
plt.plot(x, y, 'g--')
```

- Añade el título "Función f(x)", y nombra a los ejex "x" e "y"

```python
plt.plot(x, y, 'g--')
plt.title('Función de x')
plt.xlabel('x')
plt.ylabel('y')
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, y, 'g--')
ax.set_title('Función de x')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

- Añade una leyenda que diga "f(x)"

```python
fig, ax = plt.subplots(1)
ax.plot(x, y, 'g--')
ax.set_title('Función de x')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, y, 'g--', label= 'f(x)')
ax.set_title('Función de x')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

- Crea otro array z:
    - z: Resultado de aplicar la función $g(x) = \frac{1}{2}\sin(10 x) + \cos^2(x)$

```python
def g(x):
    return 1/2*np.sin(10*x) + np.cos(x)**2
```

```python
z = g(x)
```

- Crea un gráfico de dos columnas.
    - Realiza un plot sencillo de $f(x)$ y de $g(x)$ en cada uno de ellos

```python
fig, (ax_y, ax_z) = plt.subplots(1, 2)
ax_y.plot(x, y)
ax_z.plot(x, z)
```

- Dibuja el mismo gráfico anterior pero con un estilo y color de línea diferente en cada uno.

```python
fig, (ax_y, ax_z) = plt.subplots(1, 2)
ax_y.plot(x, y, '-.k')
ax_z.plot(x, z, '--r')
```

- Aumenta el gráfico anterior de tamaño.

```python
fig, (ax_y, ax_z) = plt.subplots(1, 2, figsize=(12,5))
ax_y.plot(x, y, '-.k')
ax_z.plot(x, z, '--r')
```

- Añade títulos a los ejes de cada figura, "x", "f(x)" y "g(x)"

```python
fig, (ax_y, ax_z) = plt.subplots(1, 2, figsize=(12,5))
ax_y.plot(x, y, '-.k')
ax_z.plot(x, z, '--r')
ax_y.set_xlabel('x')
ax_z.set_xlabel('x')
ax_y.set_ylabel('f(x)')
ax_z.set_ylabel('g(x)')
```

- Guardalo en un fichero .png, .jpeg y .pdf.

```python
fig.savefig('tmp/fig.png')
```

```python
fig.savefig('tmp/fig.jpeg')
```

```python
fig.savefig('tmp/fig.pdf')
```

- Ahora superpon en la misma figura ambas funciones con una legenda para cada una ('f(x)' y 'g(x)'), un título que diga "f(x) y g(x)", y nombra a los ejes "x" e "y".

```python
fig, ax = plt.subplots(1)
ax.plot(x, y, '-.k', label='f(x)')
ax.plot(x, z, '--r', label='g(x)')
ax.set_title('f(x) y g(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

## Distributions

- Crea dos arrays
    - normal: Un array de 1000 números aleatorios usando una distribución normal
    - chi: Un array de 1000 números aleatorios usando una distribución $\chi^2$ con 5 grados de libertad.

```python
normal = np.random.normal(size=1000)
chi = np.random.chisquare(5, size=1000)
```

- Crea un scatter plot con los array anteriores.

```python
fig, ax = plt.subplots(1)
ax.scatter(normal, chi)
ax.set_title('Distributions')
ax.set_xlabel('Normal')
ax.set_ylabel('Chi2')
```

- Crea un subplot de dos filas, y pinta en cada una de ellas un histograma para cada una de las distribuciones. Añade título y etiqueta los ejes como consideres.

```python
fig, (ax_1, ax_2) = plt.subplots(2, 1, sharex=True)
ax_1.hist(normal, bins=int(np.sqrt(1000)), color='g')
ax_2.hist(chi, bins=int(np.sqrt(1000)), color='m')
ax_1.set_title('Normal')
ax_2.set_title('Chi2')
ax_2.set_xlabel('Observación')
ax_1.set_ylabel('# Counts')
ax_2.set_ylabel('# Counts')
```

- Ahora superpon ambas distribuciones en un único plot

```python
fig, ax = plt.subplots(1)
ax.hist(normal, bins=int(np.sqrt(1000)), color='g', label='Normal')
ax.hist(chi, bins=int(np.sqrt(1000)), color='m', alpha=0.8, label='Chi2')
ax.set_title('Normal y Chi2')
ax.set_xlabel('Observación')
ax.set_ylabel('# Counts')
ax.legend()
```

## Pandas

- A partir del siguiente dataframe

```python
s = pd.DataFrame(np.random.randn(100, 4).cumsum(axis=0)+100,
              index=np.arange(0, 100),
             columns=['S1', 'S2', 'S3', 'S4'])
s
```

- Pinta todas las series en un úncio plot.

```python
s.plot()
```

- Ahora pintalas en plots separados

```python
s.plot(subplots=True)
```

- Ahora pinta en un mismo gráfico una de las series anteriores y añade el volumen dado por la serie siguiente

```python
s['vol'] = pd.Series(np.random.rand(100), index = np.arange(100))*100
```

```python
fig, ax = plt.subplots(1)
s['S1'].plot(ax=ax)
s['vol'].plot.bar(ax=ax)
```

```python
fig, ax = plt.subplots(1)
s['S1'].plot(ax=ax)
ax.bar(s.index, s['vol'])
```

- Mejor separalos en dos plots

```python
fig, (ax_1, ax_2) = plt.subplots(2, 1, sharex=True)
s['S1'].plot(ax=ax_1)
ax_2.bar(s.index, s['vol'])
ax_1.set_xticklabels([])
ax_1.set_xticks([])
plt.tight_layout()
```
